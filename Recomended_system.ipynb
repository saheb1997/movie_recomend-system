```python
import ast
```

```python
def convert(obj):
   L=[]
   for i in ast.literal_eval(obj) :
       L.append(i['name'])
   return L 
```

```python
def convert2(obj):
   L=[]
   for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
           L.append(i['name'])
           break
   return L 
```

```python
def convert3(obj):
   L=[]
   counter=0
   for i in ast.literal_eval(obj) :
       if(counter!=3):
           L.append(i['name'])
           counter+=1
       else:
           break
   return L 
```

```python
def convert4(string):
        return string.split()
    
```

```python
def remove_space(s):
     l=[]
     for i in s:
        l.append(i.replace(" ", ""))
     return l;
     

```

```python
#here we take only dictonary value of name
movies['genres']=movies['genres'].apply(convert)

#for the cast we are the first three character 
movies['cast']=movies['cast'].apply(convert3)

#for keywords also we do the same
movies['keywords']=movies['keywords'].apply(convert)

movies['crew']=movies['crew'].apply(convert2)

movies['overview']=movies['overview'].apply(convert4)

```

```python
movies['genres']=movies['genres'].apply(remove_space)
movies['cast']=movies['cast'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
```

5> But problem is that if we sort the similarity matix then then we lose the indexes.
6> so we  use enumerate function it makes the a tuplle of index and tuple and now we can sort and based on index x[1] and fetch the first five values.

```python
val=list(enumerate(similarity[0]))
```

```python
for i in sorted(val,reverse=True,key=lambda x:x[1])[1:6]:
    print(i[0])
```

```python
def recomended(movies):
    movie_index=new_df[new_df['title']==movies].index[0]
    print(type(movie_index))
    distance=similarity[movie_index]
    
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```
